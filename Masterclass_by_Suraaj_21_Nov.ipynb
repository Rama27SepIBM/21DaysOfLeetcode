```python
#statewise - sales
df.groupby(['State'])['Sales'].sum()
```

```python
#sort it in descending order and get the first value to get highest value
df.groupby(['State'])['Profit'].sum().sort_values(ascending=False).head(1)
```

```python
#sort it in descending order and get the last value to get lowest value
df.groupby(['State'])['Profit'].sum().sort_values(ascending=False).tail(1)
```
